```python
#   
# c.Lang = "en"
# c.Store_object = True
# c.Limit = 1000
# c.Pandas = True

# # c.Search = "#StopWar"
# # c.Since = '2022-02-25'
# # c.Until = '2022-02-27'


# # twint.run.Search(c)
# # df = twint.storage.panda.Tweets_df

# # print(df)
# # print ("this is d", d)
# #print(tlist)

def tweet_twint(hashtag, day):
  
  import twint
  from datetime import datetime, timedelta
  import nest_asyncio
  nest_asyncio.apply()
  import pandas

  c = twint.Config()
  c.Search = hashtag
  c.Pandas = True

  date = datetime.strptime(day, "%Y-%m-%d")
  modified_date = date + timedelta(days=1)

  c.Until = datetime.strftime(modified_date, "%Y-%m-%d")
  c.Limit = 2000
  
  twint.run.Search(c)
  df = twint.storage.panda.Tweets_df
  print(df)

  file_name = hashtag[1:]+'_'+ day+".csv"

  df.to_csv(file_name)

  return df

```

```python
# Give Input as hastag and date for which you want to extract tweets. Make sure that the date is older than todays date by 7 days as this code is for historic data scraping.
tweet_twint("#ukraineconflict", '2022-04-16')
```
